- Motivación intuitiva de la definición de probabilidad. Propiedades básicas de conjunto. Espacio muestral. Ejemplos. Sucesos. Definición axiomática. Propiedades de la probabilidad.
- Probabilidad condicional. Sucesos independientes. Fórmula de Bayes. Reglas multiplicativas.

- Variables aleatorias. Ejemplos. Clasificación. Función de distribución. Definición. Propiedades. Problema de muestreo con y sin reposición. Distribución empírica. Distribución binomial e hipergeométrica. Descomposición de binomial en el contexto de repeticiones independientes. Distribución geométrica, binomial negativa y Poisson.
- Distribuciones absolutamente continuas. Densidad de probabilidad. Distribución normal, ejemplos, uso de tablas.Distribuciones de Cauchy, Uniforme, Exponencial.
- Caracterización de independencia y fórmula de la convolución para suma de variables aleatorias independientes.

- Valor esperado. Definición. Propiedades. Ejemplos. Ley de Grandes Números.

- Simulación de distribuciones de VA. Método de Monte Carlo. Varianza, covarianza, coeficiente de correlación.

- Estimaciones de parámetros. Máxima verosimilitud y método de los momentos. Ejemplos. Definición de mediana y cuantiles. Estimación de estos parámetros.
- Estimación de parámetros de una distribución de Cauchy.

- Teorema Central del Límite. Aplicación a encuestas, a aproximación de Binomial y Poisson. Refinamiento de Monte Carlo.
- Intervalos de Confianza. Definición. Construcción de intervalos exactos (variables normales). Construcción aproximada usando el TCL.

- Pruebas de Hipótesis. Planteo general del problema. Región crítica. Tipos de errores. Estadística paramétrica y no paramétrica. Definición y cálculo del p-valor.
- Tests Paramétricos. Test de aleatoriedad: test de rachas. Test de correlación de rangos de Spearman. Enunciado del Teorema de Glivenko-Cantelli.
- Tests no paramétricos. Test de Kolomogorov Smirnov. Test de Lilliefors para normales y exponenciales.

- Histogramas. Estimación de densidades. Test de comparación de Kolmogorov-Smirnov para muestras independientes.
